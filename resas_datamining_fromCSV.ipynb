```python
industry_ratio = pd.read_csv('data/industry-power_20160129/01_トップ画面/稼ぐ力分析_トップ画面_市区町村_業種中分類.csv', encoding='sjis')
```

```python
city_codes = industry_ratio['市区町村コード'].unique()
```

```python
industry_abs = pd.read_csv('data/industry-power_20160129/02_グラフ分析/稼ぐ力分析_グラフ分析_市区町村_業種中分類.csv', encoding='sjis')
industry_abs = industry_abs[industry_abs['集計年']==2012]
industry_abs.index = industry_abs['市区町村コード']
industry_abs = industry_abs.sort_index()
industry_abs[-3:]
```

```python
ind_cd = industry_abs['業種中分類コード'].unique()
focus_column = ['付加価値額（企業単位）', '従業者数（事業所単位）', '従業者数（企業単位）']
for i in ind_cd:
    for f in focus_column:
        tmp_list = []
        values = industry_abs[industry_abs['業種中分類コード']==i][f].values
        for v in values: # nan 判定があるので１行で書かない
            if type(v)==str:#todo 他の場合あるのか？
                tmp_list.append(np.nan)
            else:
                tmp_list.append(v)
        all_pd[str(i)+'_'+str(f)] = tmp_list
            
```

```python
industry_sales = pd.read_csv('data/municipality-surplus_20160129/黒字赤字企業比率_市区町村_業種中分類.csv')
industry_sales2012 = industry_sales[industry_sales['集計年']==2012]
industry_sales2012.index = industry_sales2012['市区町村コード']
industry_sales2012.sort_index()
industry_sales2012[-3:]

# これが一番汎用的　todo こいつベースでリファクタリング
tmp_pd = ''
pd_data = industry_sales2012
cd = '業種中分類コード'
ind_cd = pd_data[cd].unique()
focus_column = ['赤字企業割合', '黒字企業割合']
for i in ind_cd:
    for f in focus_column:
        data = pd_data[pd_data[cd]==i][f]
        tmp_list = []
        for v in data.values: # nan 判定があるので１行で書かない
            if type(v)==str:#todo 他の場合あるのか？
                tmp_list.append(np.nan)
            else:
                tmp_list.append(v)
        tmp_tmp_pd = pd.DataFrame(tmp_list, index=data.index, columns=[str(i)+'_'+str(f)])
        if type(tmp_pd) ==str:
            tmp_pd = tmp_tmp_pd
        else:
            tmp_pd = tmp_pd.join(tmp_tmp_pd)
all_pd = all_pd.join(tmp_pd)
```
